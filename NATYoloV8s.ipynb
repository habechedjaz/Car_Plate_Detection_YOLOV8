```python
import os
import cv2
import torch
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Vérifier si un GPU est disponible et activer la précision float16 pour économiser la mémoire
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Charger le modèle YOLOv8 entraîné
model = YOLO("/content/drive/MyDrive/Nat/Test-training/weights/best.pt").to(device)
model.fuse()  # Optimisation pour accélérer l'inférence

# Récupérer la liste des classes du modèle
class_names = model.names  # Liste des noms de classes

# Dossier contenant les images à traiter
image_dir = "/content/drive/MyDrive/TEST"

# Dossier de sauvegarde des résultats
output_dir = "/content/drive/MyDrive/inférenceX"
os.makedirs(output_dir, exist_ok=True)  # Crée le dossier s'il n'existe pas

def inference_yolo_on_folder(image_dir, output_dir, conf_threshold=0.5, max_display=5):
    """
    Effectue l'inférence YOLOv8 sur toutes les images d'un dossier et sauvegarde les résultats.

    Args:
        image_dir (str): Dossier contenant les images à traiter.
        output_dir (str): Dossier où sauvegarder les images avec détection.
        conf_threshold (float): Seuil de confiance minimum pour afficher une détection.
        max_display (int): Nombre maximum d'images à afficher sur Colab.
    """
    image_list = [f for f in os.listdir(image_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

    # Vérifier si on a trouvé des images
    if not image_list:
        print("❌ Aucune image trouvée dans le dossier.")
        return

    displayed = 0  # Compteur pour affichage sur Colab

    for image_name in image_list:
        image_path = os.path.join(image_dir, image_name)
        img = cv2.imread(image_path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Inférence avec YOLOv8
        results = model.predict(img_rgb, conf=conf_threshold, device=device, half=True, imgsz=640, batch=1)[0]

        # Dessiner les boîtes détectées avec les bonnes classes
        for box in results.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            confidence = box.conf[0].item()
            class_id = int(box.cls[0].item())  # ID de la classe prédite

            if confidence > conf_threshold:
                # Nom de la classe prédite
                class_name = class_names[class_id]

                # Dessiner la boîte et l'étiquette
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                label = f"{class_name} {confidence:.2f}"
                cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Sauvegarder l'image annotée
        output_path = os.path.join(output_dir, image_name)
        cv2.imwrite(output_path, img)

        # Afficher quelques résultats sur Colab
        if displayed < max_display:
            plt.figure(figsize=(8, 6))
            plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            plt.axis("off")
            plt.title(f"Résultat pour {image_name}")
            plt.show()
            displayed += 1

    print(f"✅ Inférence terminée ! Images sauvegardées dans {output_dir}")

# Exécuter l'inférence sur tout le dossier
inference_yolo_on_folder(image_dir, output_dir)
```
